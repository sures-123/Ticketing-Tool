```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
```

```python
# Sample dataset (replace this with your actual dataset)
data = pd.read_excel("/content/Support Tracker (9).xlsx")

df = pd.DataFrame(data)
df
```

**Based on Subject, Corporation Type, Corporation, Query assigned to respective desks**

```python
df=df.dropna(axis=1, how="all")
# Split the dataset into training and testing sets
train_data, test_data = train_test_split(df, test_size=0.2, random_state=42)

# Define the features (X) and target (y)
X_train = train_data[['Subject','Corporation Type','Corporation', 'Query','Priority']]
y_train = train_data['Assigned to']
X_test = test_data[['Subject','Corporation Type','Corporation', 'Query','Priority']]
y_test = test_data['Assigned to']

X_train=X_train.dropna()
y_train = y_train.loc[X_train.index]

# Create a pipeline with CountVectorizer (you can also use TfidfVectorizer) and Multinomial Naive Bayes
model = make_pipeline(CountVectorizer(), MultinomialNB())

# Train the model
model.fit(str(X_train['Subject']) + ' ' + X_train['Corporation Type']+' '+X_train['Corporation'] + ' ' + X_train['Query']+' '+X_train['Priority'], y_train)
X_test=X_test.dropna()
y_test = y_test.loc[X_test.index]

# Make predictions on the test set
predictions = model.predict(str(X_test['Subject']) + ' ' + X_test['Corporation Type']+' '+X_test['Corporation'] + ' ' + X_test['Query']+' '+X_test['Priority'])
#print("\n################ predictions ###############\n",predictions)

# Evaluate the accuracy
accuracy = accuracy_score(y_test, predictions)
print(f'Accuracy: {accuracy}')

# Example usage: predict the "AssignedTo" for a new instance
new_instance = pd.DataFrame({'Subject':['Ram Hotels Inc'], 'Corporation Type':['Software'], 'Corporation': ['RAR'], 'Query': ['Daily Sales Follow up'], 'Priority':['Low']})
predicted_assigned_to = model.predict(new_instance['Corporation'] + ' ' + new_instance['Query'])
print(f'Predicted AssignedTo: {predicted_assigned_to[0]}')
```

```python
# Filter rows where 'Corporation' is not null
df = df[df['Corporation'].notna()]

# Filter rows where 'Subject' is not null and is of type string
df = df[df['Subject'].notna() & df['Subject'].apply(lambda x: isinstance(x, str))]

X = df['Subject']
y = df['Corporation']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Create a pipeline with CountVectorizer and Multinomial Naive Bayes
model = make_pipeline(CountVectorizer(), SVC())

# Train the model
model.fit(X_train, y_train)

# Make predictions on the test set
predictions = model.predict(X_test)
#print("\n################ predictions ###############\n", predictions)

# Evaluate the accuracy
accuracy = accuracy_score(y_test, predictions)
print(f'Accuracy: {accuracy}')

# Example usage: predict the "AssignedTo" for a new instance
new_instance = pd.DataFrame({'Subject': ['Southern Hospitality Management (SHMD) - TownePlace Suites Orlando Airport - Configuration & Automation']})
predicted_assigned_to = model.predict(new_instance['Subject'])
print(f'Predicted AssignedTo: {predicted_assigned_to[0]}')
```
